```python
from torchvision.datasets import CIFAR100

(cifar_train, cifar_test) = [
    CIFAR100(root="data",
             train=train,
             download=True)
    for train in [True, False]
]
```

```python
import torch
from torchvision.transforms import ToTensor
from torch.utils.data import TensorDataset

transform = ToTensor()

cifar_train_X = torch.stack([transform(x) for x in cifar_train.data])
cifar_test_X = torch.stack([transform(x) for x in cifar_test.data])

cifar_train = TensorDataset(
    cifar_train_X,
    torch.tensor(cifar_train.targets)
)

cifar_test = TensorDataset(
    cifar_test_X,
    torch.tensor(cifar_test.targets)
)
```

```python
from torch.utils.data import DataLoader, random_split

max_num_workers = 0

class SimpleDataModule:
    def __init__(self, train_ds, test_ds, validation=0.2, batch_size=128, num_workers=0):
        val_size = int(len(train_ds) * validation)
        train_size = len(train_ds) - val_size

        self.train_ds, self.val_ds = random_split(train_ds, [train_size, val_size])
        self.test_ds = test_ds

        self.batch_size = batch_size
        self.num_workers = num_workers

    def train_dataloader(self):
        return DataLoader(self.train_ds, batch_size=self.batch_size, shuffle=True)
```

```python
cifar_dm = SimpleDataModule(
    cifar_train,
    cifar_test,
    validation=0.2,
    num_workers=max_num_workers,
    batch_size=128
)
```

```python
for idx, (X_, Y_) in enumerate(cifar_dm.train_dataloader()):
    print('X:', X_.shape)
    print('Y:', Y_.shape)

    if idx >= 1:
        break
```
